```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
import os
```

```python
name = 'sachin'
files = os.listdir('./data/'+ name)
# print files
faces_paths = []
labels = []

for f in files:
#     ims = os.listdir('./data/' + name + "/" + f)
#     # print f, ims
#     for px in ims:
    faces_paths.append('./data/' + name + '/' + f)
    labels.append(name)

print faces_paths
print labels
```

```python
imgs = []

for fx in faces_paths:
    im = cv2.imread(fx)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    imgs.append(im)

print len(imgs)
```

```python
all_faces = []
# plt.figure(0)
for i in range(len(imgs)):
    im = imgs[i][250:500, 340:630 :]
    print im.shape
    img = cv2.resize(im,(100, 100))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    print img.shape
    plt.imshow(img, cmap='gray')
    plt.show()

    all_faces.append(img)
```

```python
print len(all_faces)
i = len(imgs)
all_faces = np.asarray(all_faces[:i])
print all_faces.shape
```

```python
lab = np.asarray(labels[:i])
print lab.shape

np.save('./dataset/' + name + '_labels', lab)
np.save('./dataset/' + name + '_data', all_faces)
```
